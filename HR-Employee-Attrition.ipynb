```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the CSV file
data = pd.read_csv(r"C:\Users\mohan reddy\OneDrive\Desktop\New folder (3)\New Folder\New Folder\HR-Employee-Attrition.csv")

# Step 3: Display first few rows
print("🔹 First 5 Rows:")
print(data.head())

# Step 4: Basic information about the dataset
print("\n🔹 Dataset Info:")
print(data.info())

# Step 5: Check for missing values
print("\n🔹 Missing Values per Column:")
print(data.isnull().sum())

# Step 6: Fill or Handle Missing Values
# Example: fill numeric columns with mean, and categorical with "Unknown"
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna("Unknown", inplace=True)
    else:
        data[column].fillna(data[column].mean(), inplace=True)

print("\n✅ Missing values handled successfully!")

# Step 7: Remove Duplicates (if any)
data.drop_duplicates(inplace=True)
print("\n✅ Duplicates removed successfully!")

# Step 8: Summary Statistics
print("\n🔹 Summary Statistics:")
print(data.describe())

# Step 9: Simple Visualization Examples
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Value Heatmap (after cleaning)")
plt.show()

# Step 10: Save the cleaned dataset
data.to_csv("cleaned_sales_data.csv", index=False)
print("\n💾 Cleaned dataset saved as 'cleaned_sales_data.csv'")
```

# Handle Missing Values

```python
# Fill numeric columns with mean, categorical with "Unknown"
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna("Unknown", inplace=True)
    else:
        data[column].fillna(data[column].mean(), inplace=True)

print("\n✅ Missing values filled successfully!")
```

# Attrition Count

```python
attrition_count = data['Attrition'].value_counts()
print(attrition_count)

sns.countplot(x='Attrition', data=data)
plt.title("Employee Attrition Count")
plt.show()
```

# Department-Wise Attrition

```python
dept_attrition = data.groupby('Department')['Attrition'].value_counts(normalize=True).unstack() * 100
print(dept_attrition)

dept_attrition.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Attrition Rate by Department (%)")
plt.ylabel("Percentage")
plt.show()
```

# Education Field vs Attrition

```python
sns.countplot(x='EducationField', hue='Attrition', data=data)
plt.title("Attrition by Education Field")
plt.xticks(rotation=45)
plt.show()
```

# Average Monthly Income by Attrition

```python
income_by_attrition = data.groupby('Attrition')['MonthlyIncome'].mean()
print(income_by_attrition)

sns.barplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title("Average Monthly Income by Attrition")
plt.show()
```

# Years at Company vs Attrition

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=data)
plt.title("Years at Company vs Attrition")
plt.show()
```

# Correlation Heatmap

```python
numeric_data = data.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
